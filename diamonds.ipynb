```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df=pd.read_csv("diamonds.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df['color'].unique()
```

```python
df.drop_duplicates(subset=['clarity'])
```

```python
df.isnull().sum()
```

```python
text=' '.join(df['cut'])
```

```python
wordcloud=WordCloud(width=800, height=400, colormap='magma',background_color='white').generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
df.boxplot(column='price', by='cut')
plt.xlabel('cut')
plt.ylabel('price')
```

```python
plt.figure(figsize=(10,5))
sns.countplot(df, x='color', order=sorted(df['color'].unique()))
plt.xlabel('color')
plt.ylabel('count')
plt.title('color count')
plt.show()
```
